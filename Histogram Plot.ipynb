```python
#pip install "Library" - if a library is missing
import matplotlib
import collections
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
arr =[] # creating an empty array and insert to it the list of each output
for i in range(50):
    with open('./Random_'+str(i)+'.out') as output:
        current = list(output)
        for j in range(len(current)-3):
            arr.append(float(current[j+3]))
```

```python
#Count of the number of appearances of each result
appear = Counter(arr)
appear
```

```python
sns.kdeplot(arr, bw_adjust=4)
plt.xlim(-0.6,0.6)
plt.title('Random Walk histogram')
plt.ylabel('Amount of Particles')
plt.xlabel('END Point')
plt.savefig('kdeplot.png')
plt.show()
```

```python
sns.histplot(arr,binwidth=0.1)
plt.xlim(-0.6,0.6)
plt.title('Random Walk histogram')
plt.ylabel('Amount of Particles')
plt.xlabel('END Point')
plt.savefig('histogram plot.png')
plt.show()
```
